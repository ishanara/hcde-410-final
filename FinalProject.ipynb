Before dividing into the dataanalytics involed in this project, I first worked to load the complete dataset into the drug_addiction variable. From my initial findings, there are 9202 rows in this dataset. 

```python
import csv

def prepare_datasets(file_path):
    """ 
    Accepts: path to a tab-separated plaintext file
    Returns: a list containing a dictionary for every row in the file, 
        with the file column headers as keys
    """
    
    with open(file_path) as infile:
        reader = csv.DictReader(infile, delimiter=',')
        list_of_dicts = [dict(r) for r in reader]
        
    return list_of_dicts
```

### Analyzing Sex and its correlation with Drug Overdoses 
In the following code block, I calulate the percentage of biological Males and Females that have died of drug overdoses in the state of Connecticut from 2012-2021. 

```python
biologicalsex_list = []

for x in drug_addiction:
    biologicalsex_list.append(x['Sex'])

countFemale = 0; 
countMale = 0; 
total = 0;

for y in biologicalsex_list: 
    total = total + 1
    if(y == 'Female'):
        countFemale = countFemale + 1
    if(y == 'Male'):
       countMale = countMale + 1 

print("Male Proportion")
males = (countMale/total)*100
print(males)
       
print("Female Proportion")
females = (countFemale/total)*100
print(females)

print(len(biologicalsex_list))
```

### Analyzing Race and its correlation with Drug Overdoses 
In the following code block I start off by gathering the unique values in the Race column in the dataset. After this point I calculated the percentage out of total dataset of each race.

```python
Race_list = []

for x in drug_addiction:
    Race_list.append(x['Race'])

unique_race_list = []

for y in Race_list:
    if(unique_race_list.count(y) == 0):
        unique_race_list.append(y)

print(unique_race_list)
print(len(unique_race_list))

print(len(Race_list))

print("_____")

BlackCount = 0
WhiteCount = 0
OtherCount = 0
AsianIndianCount = 0
IndigenousCount = 0
AsianOtherCount = 0
HawaiinCount = 0
IndigenousOther = 0
ChineseCount = 0
KoreanCount = 0

for y in Race_list:
    if(y == 'Black' or y == 'Black or African American'):
        BlackCount = BlackCount + 1
    elif(y == 'White'):
        WhiteCount = WhiteCount + 1
    elif(y == 'Other' or y == '' or y == 'Unknown'):
        OtherCount = OtherCount + 1
    elif(y == 'Asian Indian'):
        AsianIndianCount = AsianIndianCount + 1
    elif(y == 'American Indian or Alaska Native'):
        IndigenousCount = IndigenousCount + 1
    elif(y == 'Asian, Other' or y == 'Other Asian'):
        AsianOtherCount = AsianOtherCount + 1
    elif(y == 'Hawaiian'):
        HawaiinCount = HawaiinCount + 1
    elif(y == 'Native American, Other'):
        IndigenousOther = IndigenousOther + 1
    elif(y == 'Chinese'):
        ChineseCount = ChineseCount + 1
    elif(y == 'Korean'):
        KoreanCount = KoreanCount + 1
        
print((BlackCount/9202) *100)
print((WhiteCount/9202) *100)
print((OtherCount/9202) *100)
print((AsianIndianCount/9202) *100)
print((IndigenousCount/9202) *100)
print((AsianOtherCount/9202) *100)
print((HawaiinCount/9202) *100)
print((IndigenousOther/9202) *100)
print((ChineseCount/9202) *100)
print((KoreanCount/9202) *100)      
```
